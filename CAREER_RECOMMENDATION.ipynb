```
import ipywidgets as widgets
from IPython.display import display, clear_output

# ---------------- TITLE ----------------
title = widgets.HTML(
    value="<h2 style='color:blue;'>💼 Career Recommendation System</h2>"
)

# ---------------- INTEREST OPTIONS ----------------
interest_options = [
    "Coding", "Math", "Design", "Biology", "Writing",
    "Business", "Teaching", "Gaming", "Art", "Communication",
    "Psychology", "Finance", "Law", "Music", "Sports",
    "Cybersecurity", "AI", "Healthcare", "Management", "Media"
]

# ---------------- WIDGETS ----------------
interest1 = widgets.Dropdown(
    options=interest_options,
    description='1st Interest:',
    layout=widgets.Layout(width='450px'),
    style={'description_width': 'initial'}
)

interest2 = widgets.Dropdown(
    options=interest_options,
    description='2nd Interest:',
    layout=widgets.Layout(width='450px'),
    style={'description_width': 'initial'}
)

interest3 = widgets.Dropdown(
    options=interest_options,
    description='3rd Interest:',
    layout=widgets.Layout(width='450px'),
    style={'description_width': 'initial'}
)

skill_slider = widgets.IntSlider(
    value=5,
    min=1,
    max=10,
    step=1,
    description='Skill Level:',
    layout=widgets.Layout(width='450px'),
    style={'description_width': 'initial'}
)

button = widgets.Button(
    description="Recommend Career",
    button_style='info',
    icon='briefcase'
)

output = widgets.Output()

# ---------------- CAREER LOGIC ----------------
career_map = {
    "Coding": ["Junior Programmer", "Web Developer", "Software Engineer"],
    "Math": ["Accountant", "Data Analyst", "Actuary"],
    "Design": ["Design Assistant", "Graphic Designer", "UI/UX Designer"],
    "Biology": ["Lab Assistant", "Biotechnologist", "Doctor"],
    "Writing": ["Blogger", "Content Writer", "Journalist"],
    "Business": ["Sales Executive", "Marketing Manager", "Entrepreneur"],
    "Teaching": ["Tutor", "Teacher", "Professor"],
    "Gaming": ["Game Tester", "Game Developer", "Game Designer"],
    "Art": ["Sketch Artist", "Graphic Designer", "Animator"],
    "Communication": ["Receptionist", "HR Manager", "Public Relations Manager"],
    "Psychology": ["Counselor", "Therapist", "Psychologist"],
    "Finance": ["Bank Clerk", "Accountant", "Investment Banker"],
    "Law": ["Paralegal", "Legal Advisor", "Lawyer"],
    "Music": ["Performer", "Music Producer", "Singer"],
    "Sports": ["Coach Assistant", "Fitness Trainer", "Athlete"],
    "Cybersecurity": ["Security Intern", "Cybersecurity Analyst", "Ethical Hacker"],
    "AI": ["AI Intern", "ML Engineer", "AI Researcher"],
    "Healthcare": ["Care Assistant", "Nurse", "Physiotherapist"],
    "Management": ["Team Lead", "Project Manager", "Operations Manager"],
    "Media": ["Content Creator", "Digital Creator", "Media Strategist"]
}

# ---------------- FUNCTION ----------------
def get_level(skill):
    if skill >= 8:
        return 2
    elif skill >= 5:
        return 1
    else:
        return 0

def on_click(b):
    with output:
        clear_output()

        i1 = interest1.value
        i2 = interest2.value
        i3 = interest3.value
        skill = skill_slider.value
        level = get_level(skill)

        # Main recommendation based on first interest
        main_career = career_map[i1][level]

        # Additional recommendations
        second_career = career_map[i2][level]
        third_career = career_map[i3][level]

        print("Top Interests Chosen:")
        print("1.", i1)
        print("2.", i2)
        print("3.", i3)
        print("\nSkill Level:", skill, "/10")

        print("\n🎯 Recommended Careers:")
        print("1.", main_career)
        print("2.", second_career)
        print("3.", third_career)

button.on_click(on_click)

# ---------------- DISPLAY ----------------
display(
    title,
    interest1,
    interest2,
    interest3,
    skill_slider,
    button,
    output
)
```
