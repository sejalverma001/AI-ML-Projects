```
# Food Calorie Estimator - Interactive Colab Version (Upgraded)

import pandas as pd
from sklearn.tree import DecisionTreeRegressor
import ipywidgets as widgets
from IPython.display import display, clear_output

# ---------------- DATASET ----------------
data = {
    "Food": [
        "Apple","Banana","Orange","Mango","Grapes",
        "Rice","Brown Rice","Bread","Roti","Pasta",
        "Egg","Chicken Breast","Fish","Paneer","Tofu",
        "Milk","Curd","Cheese","Butter",
        "Burger","Pizza","Fries","Sandwich","Noodles",
        "Salad","Oats","Cornflakes","Ice Cream","Chocolate"
    ],
    "Weight": [
        100,100,100,100,100,
        100,100,50,40,100,
        50,100,100,100,100,
        200,100,30,10,
        150,120,100,120,150,
        100,40,30,100,50
    ],
    "Calories": [
        52,89,47,60,69,
        130,111,133,120,157,
        78,165,206,265,144,
        124,98,120,72,
        354,266,312,250,220,
        33,150,110,207,270
    ]
}

df = pd.DataFrame(data)

# ---------------- ENCODING ----------------
food_map = {name: i for i, name in enumerate(df["Food"].unique())}
df["FoodCode"] = df["Food"].map(food_map)

X = df[["FoodCode", "Weight"]]
y = df["Calories"]

# ---------------- MODEL ----------------
model = DecisionTreeRegressor()
model.fit(X, y)

# ---------------- WIDGETS ----------------
title = widgets.HTML(
    value="<h2 style='color:green;'>🍎 Food Calorie Estimator</h2>"
)

search_box = widgets.Text(
    placeholder='Type food name...',
    description='Search:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='400px')
)

food_dropdown = widgets.Dropdown(
    options=list(food_map.keys()),
    description='Food:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='400px')
)

weight_input = widgets.IntText(
    value=100,
    description='Weight (g):',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='250px')
)

button = widgets.Button(
    description="Estimate Calories",
    button_style='success',
    icon='check'
)

output = widgets.Output()

# ---------------- FUNCTIONS ----------------
def update_options(change):
    text = search_box.value.lower()
    filtered = [food for food in food_map.keys() if text in food.lower()]
    food_dropdown.options = filtered if filtered else list(food_map.keys())

search_box.observe(update_options, names='value')

def on_click(b):
    with output:
        clear_output()
        food_name = food_dropdown.value
        weight = weight_input.value

        code = food_map[food_name]
        input_data = pd.DataFrame([[code, weight]], columns=["FoodCode", "Weight"])
        prediction = model.predict(input_data)[0]

        # calorie level
        if prediction < 100:
            level = "Low Calorie ✅"
        elif prediction < 250:
            level = "Medium Calorie ⚡"
        else:
            level = "High Calorie 🔥"

        print("Food:", food_name)
        print("Weight:", weight, "grams")
        print("Estimated Calories:", round(prediction, 2), "kcal")
        print("Category:", level)

button.on_click(on_click)

# ---------------- DISPLAY ----------------
display(
    title,
    search_box,
    food_dropdown,
    weight_input,
    button,
    output
)
```
